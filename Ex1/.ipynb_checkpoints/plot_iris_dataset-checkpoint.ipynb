
# The Iris Dataset

This data sets consists of 3 different types of irises'
(Setosa, Versicolour, and Virginica) petal and sepal
length, stored in a 150x4 numpy.ndarray

The rows being the samples and the columns being:
Sepal Length, Sepal Width, Petal Length and Petal Width.

The below plot uses the first two features.
See `here <https://en.wikipedia.org/wiki/Iris_flower_data_set>`_ for more
information on this dataset.



```python
print(__doc__)

import matplotlib.pyplot as plt
import sklearn
import random
import numpy as np
from numpy import asarray
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import VotingClassifier
```

```python
# Load data
iris = datasets.load_iris()

X = iris.data[:, :2] # Features in X
y = iris.target      # Lables in y

#Devide into train and validate at 1:3 ratio
x_val, x_train, y_val, y_train = [],[],[],[]

i = 0
while i < len(X):
    if i % 3 != 0:
        x_train.extend([X[i]])
        y_train.extend([y[i]])
    else:
        x_val.extend([X[i]])
        y_val.extend([y[i]])
    i += 1
        
neigh1 = KNeighborsClassifier(n_neighbors=1)
neigh1.fit(x_train, y_train)
pred1 = neigh1.predict(x_val)

neigh5 = KNeighborsClassifier(n_neighbors=5)
neigh5.fit(x_train, y_train)
pred5 = neigh1.predict(x_val)

neigh10 = KNeighborsClassifier(n_neighbors=10)
neigh10.fit(x_train, y_train)
pred10 = neigh1.predict(x_val)

neigh20 = KNeighborsClassifier(n_neighbors=20)
neigh20.fit(x_train, y_train)
pred20 = neigh1.predict(x_val)

neigh30 = KNeighborsClassifier(n_neighbors=30)
neigh30.fit(x_train, y_train)
pred30 = neigh1.predict(x_val)

x_train = asarray(x_train)

x_min, x_max = x_train[:, 0].min() - .5, x_train[:, 0].max() + .5
y_min, y_max = x_train[:, 1].min() - .5, x_train[:, 1].max() + .5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

f, axarr = plt.subplots(2, 3, sharex='col', sharey='row', figsize=(15, 10))

for idx, clf, tt in zip(([0,0],[0,1],[1,0],[1,1],[0,2]),
                        [neigh1, neigh5, neigh10, neigh20, neigh30],
                        ['k=1', 'k=5', 'k=10', 'k=20', 'k=30']):

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    axarr[idx[0], idx[1]].contourf(xx, yy, Z, alpha=0.4)
    axarr[idx[0], idx[1]].scatter(x_train[:, 0], x_train[:, 1], c=y_train, s=20, edgecolor='k')
    axarr[idx[0], idx[1]].set_title(tt)

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.figure(2, figsize=(15, 10))
plt.clf()
plt.show()

print("k=1 Training: " + str(neigh1.score(x_train,y_train)))
print("k=1 Validate: " + str(neigh1.score(x_val,y_val)))

print("k=5 Training: " + str(neigh5.score(x_train,y_train)))
print("k=5 Validate: " + str(neigh5.score(x_val,y_val)))

print("k=10 Training: " + str(neigh10.score(x_train,y_train)))
print("k=10 Validate: " + str(neigh10.score(x_val,y_val)))

print("k=20 Training: " + str(neigh20.score(x_train,y_train)))
print("k=20 Validate: " + str(neigh20.score(x_val,y_val)))

print("k=30 Training: " + str(neigh30.score(x_train,y_train)))
print("k=30 Validate: " + str(neigh30.score(x_val,y_val)))
```
